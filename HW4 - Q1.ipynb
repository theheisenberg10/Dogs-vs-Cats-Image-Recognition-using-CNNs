```python
import pandas as pd
%matplotlib inline
import math
from sklearn.model_selection import cross_val_score, train_test_split, KFold
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import GridSearchCV
#from sklearn.cross_validation import train_test_split
from sklearn.metrics import *
from sklearn.preprocessing import MinMaxScaler, StandardScaler

from sklearn import neighbors, datasets
import warnings
warnings.filterwarnings("ignore") 
```

```python
np.random.seed(100)
input_data = np.random.uniform(math.pi * -2, math.pi * 2, 120000)
```

```python
input_data.shape
```

```python
output_data = 2 * (2 * np.cos(input_data) ** 2 - 1) ** 2 - 1 
```

```python
X = input_data.reshape((-1, 1))
Y = output_data
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.5, random_state=10)
```

```python
reg = linear_model.LinearRegression()
reg.fit(x_train, y_train)
yhat = reg.predict(x_test)
mean_squared_error(y_true=y_test, y_pred=yhat)
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
import tensorflow.keras.backend as K
from tensorflow.keras.models import Sequential
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Dense
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier, KerasRegressor
from tensorflow.keras import activations
from keras.layers import Dropout
from keras.layers import BatchNormalization
```

```python
#building the model
model = Sequential()

final_score = []
neurons=np.array([24,48,72,128,256])

for i in neurons:
    model.add(Dense(i, activation='relu', input_dim = 1))
    
    model.add(Dense(i, activation='relu'))

    model.add(Dense(1, activation='linear'))
    
    #compiling the model with mean sqaured error evaluating model performance
    model.compile(optimizer='adam', loss='mean_squared_error')
    
    #train model
    model.fit(x_train, y_train, epochs=2000, batch_size = 3000, verbose = 0)
    
    #making predictions
    y_predictions = model.predict(x_test)
    
    #evaluate the model
    scores = model.evaluate(x_test, y_test)    
    final_score.append(scores)
```

```python
neurons=[24,48,72,128,256]
mydf = pd.DataFrame(final_score,neurons, columns = ['1 layer'])
mydf
```

```python
#building the model
model = Sequential()

final_score_2 = []
neurons=np.array([12,24,36])

for i in neurons:
    model.add(Dense(i, activation='relu', input_dim = 1))
    
    model.add(Dense(i, activation='relu'))
    
    model.add(Dense(i, activation='relu'))

    model.add(Dense(1, activation='linear'))
    
    #compiling the model with mean sqaured error evaluating model performance
    model.compile(optimizer='adam', loss='mean_squared_error')
    
    #train model
    model.fit(x_train, y_train, epochs=30)
    
    #making predictions
    y_predictions = model.predict(x_test)
    
    #evaluate the model
    scores = model.evaluate(x_test, y_test)
    final_score_2.append(scores)
```

```python
neurons=[24,48,72]
mydf2 = pd.DataFrame(final_score_2,neurons, columns = ['2 layer'])
mydf2
```

```python
#building the model
model = Sequential()

final_score_3 = []
nuerons=np.array([8,16,24])

for i in nuerons:
    model.add(Dense(i, activation='relu', input_dim = 1))
    
    model.add(Dense(i, activation='relu'))
    
    model.add(Dense(i, activation='relu'))
    
    model.add(Dense(i, activation='relu'))

    model.add(Dense(1, activation='linear'))
    
    #compiling the model with mean sqaured error evaluating model performance
    model.compile(optimizer='adam', loss='mean_squared_error')
    
    #train model
    model.fit(x_train, y_train, epochs=30)
    
    #making predictions
    y_predictions = model.predict(x_test)
    
    #evaluate the model
    scores = model.evaluate(x_test, y_test)
    final_score_3.append(scores)
```

```python
nuerons=[24,48,72]
mydf3 = pd.DataFrame(final_score_3,nuerons, columns = ['3 layer'])
mydf3
```

```python
final_df = mydf.join(mydf2, how = 'outer').join(mydf3, how = 'outer')
final_df
```

```python
fig, ax = plt.subplots(figsize=(10,5))
plt.title('Shallow V.S Deep Neural Network',size=18, y=1.06) 

ax.plot(final_df.index.values, final_df["1 layer"],color='black')
ax.plot(final_df.index.values, final_df["2 layer"],color='blue')
ax.plot(final_df.index.values, final_df["3 layer"],color='darkgreen')
ax.set_xlabel('Number of nuerons',fontsize=15)
ax.set_ylabel('Test MSE', color='black',fontsize=15)
ax.tick_params(axis='both', which='major', labelsize=14)
ax.tick_params('y', colors='black')
```
